## Melbourne housing market dataset

The Melbourne housing market dataset contains a wealth of information about the **properties sold in Melbourne, Australia between January 2016 and December 2018**.

The dataset includes **21 columns**, each providing different information about the properties, such as their location, size, number of rooms, type of property, and price. Some columns contain continuous numerical data, such as 'Price', 'Distance', 'Landsize', and 'BuildingArea', while others contain categorical data, such as 'Suburb', 'Type', 'Method', 'Regionname', and 'CouncilArea'.

Understanding the information contained in each column is essential for conducting **exploratory data analysis** on the dataset and gaining **insights into the Melbourne housing market**.

### Dataset columns:

- **Suburb**: The suburb where the property is located
- **Address**: The street address of the property
- **Rooms**: The number of rooms in the property
- **Type**: The type of the property (house, townhouse, unit, etc.)
- **Price**: The price at which the property was sold
- **Method**: The method of sale (auction, private treaty, etc.)
- **SellerG**: The agency or agent who sold the property
- **Date**: The date on which the property was sold
- **Distance**: The distance of the property from Melbourne's central business district (CBD) in kilometers
- **Postcode**: The postcode of the suburb where the property is located
- **Bedroom2**: The number of bedrooms in the property (other than the master bedroom)
- **Bathroom**: The number of bathrooms in the property
- **Car**: The number of car spaces in the property
- **Landsize**: The size of the land on which the property is located in square meters
- **BuildingArea**: The size of the building on the land in square meters
- **YearBuilt**: The year in which the building was constructed
- **CouncilArea**: The local government area in which the property is located
- **Lattitude**: The latitude coordinate of the property
- **Longtitude**: The longitude coordinate of the property
- **Regionname**: The general region in Melbourne (west, north, south-east, etc.)
- **Propertycount**: The number of properties in the suburb

### 9/ Does the dataset contain any properties that are priced significantly above or below the average? To investigate, create a box plot to detect potential outliers, and use Tukey's method with a 1.5 threshold to create a new dataframe without these outliers. What percentage of properties are identified as outliers using Tukey's method?

##### Box plot to visualize outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a box plot to visualize outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_mb_price['price'])
plt.title('Box Plot of Property Prices')
plt.xlabel('Property Price')
plt.show()
```

##### Calculate Tukey's limits

```python
# Calculate the quartiles and interquartile range (IQR)
Q1 = df_mb_price['price'].quantile(0.25)  # First quartile
Q3 = df_mb_price['price'].quantile(0.75)  # Third quartile
IQR = Q3 - Q1  # Interquartile range

# Calculate lower and upper limits
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

lower_bound, upper_bound
```

##### Identify outliers and its percentage

```python
# Identify outliers
outliers = df_mb_price[(df_mb_price['price'] < lower_bound) | (df_mb_price['price'] > upper_bound)]

# Calculate the percentage of outliers
percentage_outliers = (len(outliers) / len(df_mb_price)) * 100

# Percentage of properties identified as outliers using Tukey's method
print(f"Percentage of properties identified as outliers using Tukey's method: {percentage_outliers:.2f}%")
```

### 11/ What is the outcome of performing a normality test, such as the Kolmogorov-Smirnov test, on the distribution of property prices? Specifically, what possible conclusions can be drawn from the test results regarding the similarity of the distribution to a normal distribution?

##### Kolmogorov-Smirnov test

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Kolmogorov-Smirnov test
ks_statistic, p_value = stats.kstest(df_mb_price['price'], 'norm', args=(df_mb_price['price'].mean(), df_mb_price['price'].std()))

# Results
print(f"KS Statistic: {ks_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("The distribution of property prices is significantly different from a normal distribution.")
else:
    print("The distribution of property prices is not significantly different from a normal distribution.")
```

### 14/ Is there a statistically significant association between the region (Regionname) and the price of properties? Use chi-square test to assess the independence between region and price and Cramer's V coefficient to quantify the strength of the association between these two features.

##### Discretize the price in different ranges.

```python
import pandas as pd
import numpy as np

# Create price ranges
bins = [0, 500000, 1000000, 1500000, 2000000, np.inf]  # Define the bins for price ranges
labels = ['0-500k', '500k-1M', '1M-1.5M', '1.5M-2M', '2M+']  # Define the labels for the bins
df_mb_price['Price_Range'] = pd.cut(df_mb_price['price'], bins=bins, labels=labels)

# Display the first few rows to confirm the new Price_Range column
df_mb_price[['price', 'Price_Range']].head()
```

##### Chi-square test

```python
# Create a contingency table
contingency_table = pd.crosstab(df_mb_price['region_name'], df_mb_price['Price_Range'])

# Chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Results
print(f"Chi2 Stat: {chi2_stat:.2f}")
print(f"P-Value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
```

##### Cramér's V coefficent

```python
# Calculate Cramér's V
n = contingency_table.sum().sum()  # Total number of observations
phi2 = chi2_stat / n  # Calculate phi squared
r, k = contingency_table.shape  # Get the number of rows and columns
cramers_v = np.sqrt(phi2 / min(k - 1, r - 1))  # Calculate Cramér's V

# Results
print(f"Cramér's V: {cramers_v:.4f}")
```

##### **SOLUTION 14:** Yes, there is a significant relationship between the region (Regionname) and the price of properties, and the relationship is medium.

##### Count number of houses sold per month

```python
# Count the number of houses sold per month
houses_sold_by_month = df_mb_price['month'].value_counts()

# Results
print(houses_sold_by_month)
```

##### Graphic barplot

```python
# Plot the number of houses sold per month.
import matplotlib.pyplot as plt
houses_sold_by_month.plot(kind='bar', color='skyblue')
plt.title('Number of Houses Sold by Month')
plt.xlabel('Month')
plt.ylabel('Number of Houses Sold')
plt.xticks(rotation=45)
plt.show()
```
